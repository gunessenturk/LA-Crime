```python
import pandas as pd
import folium
```

```python
LA_crime = pd.read_csv('LA_Crime_Data_from_2010_to_Present.csv')
```

```python
LA_crime['Location '] = LA_crime['Location '].str.replace('\(','')
LA_crime['Location '] = LA_crime['Location '].str.replace('\)','')
```

```python
LA_crime['Latitude'] = LA_crime['Location '].str.replace(',.*', '').astype('float')
```

```python
LA_crime['Longitude'] = LA_crime['Location '].str.replace('.*,', '').astype('float')
```

```python
LA_crime['Year Occured'] = LA_crime['Date Occurred'].str.replace('.*/.*/', '').astype('int')
```

```python
LA_crime_2018 = LA_crime[LA_crime['Year Occured'] == 2018]
```

```python
dist_peace_2018 = LA_crime_2018[LA_crime_2018['Crime Code Description'] == 'DISTURBING THE PEACE'][['Latitude', 'Longitude', 'Area Name']]
```

```python
LA_map_2018 = folium.Map(location=[33.9829, -118.3338],
                        zoom_start=10,
                        tiles='cartodbpositron')
```

```python
for i in range(len(dist_peace_2018)):
    lat = dist_peace_2018['Latitude'].iloc[i]
    lon = dist_peace_2018['Longitude'].iloc[i]
    folium.CircleMarker([lat, lon], radius=1, color='red', popup=dist_peace_2018['Area Name'].iloc[i]).add_to(LA_map_2018)
```

```python
legend_html =   '''
                <div style="position: fixed; 
                            bottom: 50px; left: 50px; width: 180px; height: 25px; 
                            border:2px solid grey; z-index:9999; font-size:14px;
                            ">&nbsp; Disturbing the Peace &nbsp; <i class="fa fa-circle" style="color:red"></i><br>
                              
                              
                </div>
                ''' 

LA_map_2018.get_root().html.add_child(folium.Element(legend_html))
LA_map_2018
```

```python
LA_map_2018.save("LA_disturbing_the_peace.html")
```
